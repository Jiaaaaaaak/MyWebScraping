```python
import pyautogui
import time

"""
取得起始點的方式
1. time.sleep(2) 搭配 pyautogui.position() 取得滑鼠位置
2. 在 cmd 使用 python -m pyautogui 進入互動模式
"""

# 設定起始點
start_point = (888, 400)
# 移動滑鼠到起始點
pyautogui.moveTo(start_point[0], start_point[1])

distance = 20
# 按下滑鼠左鍵

while distance > 0 :
    # 向右拖曳
    pyautogui.dragRel(distance, 0)
    # 向下拖曳
    distance = -10
    pyautogui.dragRel(0, distance)

    # 向左拖曳
    pyautogui.dragRel(-distance, 0)

    # 向上拖曳
    distance = -10
    pyautogui.dragRel(0, -distance)
    # pyautogui.dragRel(190, 0)
    # pyautogui.dragRel(0, 180)


# 放開滑鼠左鍵
```
