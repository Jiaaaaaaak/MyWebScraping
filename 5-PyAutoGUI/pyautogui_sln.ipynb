```python
import pyautogui
import time

"""
取得起始點的方式
1. time.sleep(2) 搭配 pyautogui.position() 取得滑鼠位置
2. 在 cmd 使用 python -m pyautogui 進入互動模式
"""

start_point = (500, 350)
pyautogui.moveTo(start_point)

pyautogui.mouseDown()
distance = 200

while distance > 0:
    pyautogui.drag(distance, 0, duration=0.1)
    distance -= 10
    pyautogui.drag(0, distance, duration=0.1)
    pyautogui.drag(-distance, 0, duration=0.1)
    distance -= 10
    pyautogui.drag(0, -distance, duration=0.1)

pyautogui.mouseUp()
```

```python
import pyautogui
import keyboard
# pyautogui.PAUSE = 0.1  # 每個 PyAutoGUI 函式呼叫後暫停 1 秒

point_list = [
    (1000, 900),
    (1200, 900),
    (1400, 900),
    (1600, 900),
]

def click():
    for point in point_list:
        color = pyautogui.pixel(point[0], point[1])

        if color[0] == 0 and color[1] == 0:
            pyautogui.mouseDown(point[0], point[1])
            pyautogui.mouseUp()

while True:
    if keyboard.is_pressed('Ctrl'):
        click()
    elif keyboard.is_pressed('q'):
        break
```

```python
import keyboard
"""
https://cjdgrevival.com/
blue_rgb = (101, 189, 187)
orange_rgb = (243, 71, 40)
orange_rgb_alt = (248, 119, 0)
yellow_rgb = (243, 181, 0)
"""

# 顏色範圍設定
blue_range = {
    "r": (95, 105),
    "g": (180, 195),
    "b": (180, 190)
}

orange_range = {
    "r": (240, 250),
    "g": (70, 75),
    "b": (35, 45)
}

orange_alt_range = {
    "r": (245, 255),
    "g": (110, 125),
    "b": (0, 10)
}

yellow_range = {
    "r": (235, 245),
    "g": (175, 185),
    "b": (0, 10)
}

def in_range(color, range_dict):
    r, g, b = color
    return (range_dict["r"][0] <= r <= range_dict["r"][1] and
            range_dict["g"][0] <= g <= range_dict["g"][1] and
            range_dict["b"][0] <= b <= range_dict["b"][1])

def click():
    pixel = pyautogui.pixel(720, 590)

    if in_range(pixel, blue_range):
        pyautogui.press('D')

    elif in_range(pixel, orange_range):
        pyautogui.press('F')

    elif in_range(pixel, orange_alt_range):
        for _ in range(3):
            pyautogui.press('D')
            pyautogui.press('F')

    elif in_range(pixel, yellow_range):
        for _ in range(3):
            pyautogui.press('D')
            pyautogui.press('F')
while True:
    if keyboard.is_pressed('Ctrl'):
        click() 
    elif keyboard.is_pressed('Q'):
        print("退出程式")
        break
```
